```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids
```

```python
df = pd.read_csv("Data\\Human activity.csv").drop("subject", axis=1)
activities = df["Activity"]
inputs = df.drop("Activity", axis=1)
inputs.head()
```

```python
sns.barplot(activities, errorbar=None)
```

```python
sns.scatterplot(df, x="angle(X,gravityMean)", y="tBodyAccMag-mean()", hue="Activity")
```

```python
KMeans2 = KMeans(n_clusters=2)
KMeans2.fit(inputs)
sns.scatterplot(inputs, x="angle(X,gravityMean)", y="tBodyAccMag-mean()", hue=KMeans2.predict(inputs))
```

```python
KMeans3 = KMeans(n_clusters=3)
KMeans3.fit(inputs)
sns.scatterplot(inputs, x="angle(X,gravityMean)", y="tBodyAccMag-mean()", hue=KMeans3.predict(inputs))
```

```python
KMed2 = KMedoids(n_clusters=2)
KMed2.fit(inputs)
sns.scatterplot(inputs, x="angle(X,gravityMean)", y="tBodyAccMag-mean()", hue=KMed2.predict(inputs))
```

```python
KMed3 = KMedoids(n_clusters=3)
KMed3.fit(inputs)
sns.scatterplot(inputs, x="angle(X,gravityMean)", y="tBodyAccMag-mean()", hue=KMed3.predict(inputs))
```
