```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from keras.preprocessing import image_dataset_from_directory
```

```python
testing = image_dataset_from_directory("Brain_tumor\\Testing", pad_to_aspect_ratio=True)
training = image_dataset_from_directory("Brain_tumor\\Training", pad_to_aspect_ratio=True)
```

```python
print(testing.class_names)
print(training.class_names)
```

```python
batch = testing.take(1)
for image, label in batch:
    image_shape = image.shape
    print("Width: {}\tHeight: {}\tChannels: {}".format(image_shape[1], image_shape[2], image_shape[3]))
```

```python
i=1
fig = plt.figure()
batch2 = testing.take(6)
for image, label in batch2:
    for image2, label2 in zip(image, label):
        fig.add_subplot(2, 3, i)
        plt.imshow(image2.numpy().astype('int'))
        plt.title(testing.class_names[label2])
        i += 1
        if i > 6:
            break
    break
```
